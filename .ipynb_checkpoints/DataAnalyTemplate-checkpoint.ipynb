## 三、类别型数据分析

### 1.  类别型数据的描述统计信息

```python
def category_distribute(df,subset):
    """
    绘制Pie饼图，查看数据表分布情况
    """
    lis_dis=df[subset].value_counts()
    labels=lis_dis.index
    sns.set(font_scale=1.5)
    plt.rcParams['font.sans-serif']='SimHei'
    plt.rcParams['axes.unicode_minus']=False
    plt.figure(figsize=(15,15))
    plt.title('{}分布占比'.format(subset),fontdict={'fontsize':18})
    plt.pie(lis_dis,
                labels=labels,
                autopct='%.2f%%',
                startangle=90,
                counterclock=False,
                textprops={'fontsize':12,'color':'black'},
                colors=sns.color_palette('hls',n_colors=18)) 
    plt.legend(loc='best',shadow=True,fontsize=11)
```

```python
# 应用category_distribute函数，分析类别型数据category_col的分布情况
category_col = "neighbourhood"
category_distribute(dataframe, category_col)
```

### 4.2. 数值型数据-基础信息/分位数信息统计

```python
def num_describe(df, subsets):
    """
    探索性数据分析，数据'偏度','峰度','均值','标准差','变异系数','极差','四分位距','中位数'
    """
    data={}
    for i in subsets:
        data.setdefault(i,[])
        data[i].append(df[i].skew())
        data[i].append(df[i].kurt())
        data[i].append(df[i].mean())
        data[i].append(df[i].std())
        data[i].append(df[i].std()/df[i].mean())
        data[i].append(df[i].max()-df[i].min())
        data[i].append(df[i].quantile(0.75)-df[i].quantile(0.25))
        data[i].append(df[i].median())
    data_df=pd.DataFrame(data,index=['偏度','峰度','均值','标准差','变异系数','极差','四分位距','中位数'],columns=subsets)
    return data_df.T
```

```python
# 应用num_describe函数，分析数值型数据num_subsets的分布情况
num_describe(dataframe, num_subsets)
```

## 五、数据集频数TopN

```python
def get_TopN(df, groupCol, orderCol,num):
    result=[]
    groups=df.groupby(groupCol)
    for x,group in groups:
        if num>0:
            result.append(group.sort_values(by=orderCol,ascending=False)[:num])
        if num<0:
            result.append(group.sort_values(by=orderCol,ascending=False)[num:])
    result=pd.concat(result)
    return result
```

```python
# 统计，各地区的房源带看次数最多的100条的短租信息
# 分析：房源带看次数Top100，房源热度高
reviews_top100 = get_TopN(dataframe,'neighbourhood','number_of_reviews',100)
reviews_top100.head()
```

```python
# 统计，各地区的房源带看次数最少的100条的短租信息
# 分析：房源带看次数Bottom100，房源热度低
reviews_bottom100=get_TopN(dataframe,'neighbourhood','number_of_reviews',-100)
reviews_bottom100.head()
```
